```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
data = pd.read_csv('/content/monthly_deaths.csv', usecols=['date', 'births', 'deaths'])
data['date'] = pd.to_datetime(data['date'])

# Calculate mortality rate
data['mortality_rate'] = data['deaths'] / data['births']

# Plot mortality rates over time
plt.figure(figsize=(10, 6))
plt.plot(data['date'], data['mortality_rate'], label='Mortality Rate')
handwashing_start = pd.to_datetime('1847-06-01')
plt.axvline(handwashing_start, color='red', linestyle='--', label='Handwashing starts')
plt.xlabel('Date')
plt.ylabel('Mortality Rate')
plt.title('Mortality Rate Over Time')
plt.legend()
plt.show()

# Calculate and compare mortality rates
before_handwashing = data[data['date'] < handwashing_start]
after_handwashing = data[data['date'] >= handwashing_start]
mean_before = before_handwashing['mortality_rate'].mean()
mean_after = after_handwashing['mortality_rate'].mean()

print(f'Mean mortality rate before handwashing: {mean_before:.4f}')
print(f'Mean mortality rate after handwashing: {mean_after:.4f}')

```

```python
import seaborn as sns

avg_mortality_data = pd.DataFrame({
    'Period': ['Before Handwashing', 'After Handwashing'],
    'Mortality_Rate': [mean_before, mean_after]
})

plt.figure(figsize=(8, 6))
sns.barplot(x='Period', y='Mortality_Rate', data=avg_mortality_data)
plt.xlabel('Period')
plt.ylabel('Average Mortality Rate')
plt.title('Average Mortality Rate Before and After Handwashing')
plt.show()
```
